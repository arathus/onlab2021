```python
import pandas as pd
import json
import os
import sys; sys.path.append('/home/imetomi/Minyma/Minyma-DEV-Codebase')
from os import walk
from datetime import datetime
from dateutil.parser import parse
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
import numpy as np
```

```python
path = "../Data/lob/btc/"
_, _, filenames = next(walk(path))
filenames = sorted(filenames)
print(filenames)
```

```python
temp = json.load(open(path + filenames[0], 'r'))
```

```python
for key in temp:
    ask = np.array(temp[key]['a'], dtype=float)
    bid = np.array(temp[key]['b'], dtype=float)
    ask_vol = ask.sum(axis=0)[1]
    bid_vol = bid.sum(axis=0)[1]
    ask_price = ask.min(axis=0)[0]
    bid_price = bid.max(axis=0)[0]
    date = datetime.fromtimestamp(int(key) / 1000)
    
    sorted_ask = ask[ask[:,0].argsort()]
    start = sorted_ask[0][0]
    ask_depth = np.zeros(20)
    diff = 10
    current_vol = 0
    cnt = 0
    for i in range(len(sorted_ask)):
        if abs(start - sorted_ask[i][0]) < diff:
            current_vol += sorted_ask[i][1]
        else:
            if cnt < 20:
                ask_depth[cnt] = (current_vol)
                cnt+=1
                current_vol = 0
                start = sorted_ask[i][0]
            else:
                continue

    
    sorted_bid = bid[bid[:,0].argsort()]
    start = sorted_bid[0][0]
    bid_depth = np.zeros(20)
    diff = 10
    current_vol = 0
    cnt = 0
    for i in range(len(sorted_bid)):
        if abs(start - sorted_bid[i][0]) < diff:
            current_vol += sorted_bid[i][1]
        else:
            if cnt < 20:
                bid_depth[cnt] = (current_vol)
                cnt+=1
                current_vol = 0
                start = sorted_bid[i][0]
            else:
                continue


    print(date, int(key), ask_price, bid_price, ask_vol, bid_vol, bid_depth, ask_depth)
```

```python
ask_depth = ['ask' + str(i * 10) for i in range(1, 21)]
bid_depth = ['bid' + str(i * 10) for i in range(1, 21)]
columns = ['date', 'timestamp', 'ask', 'bid', 'askvol', 'bidvol'] + ask_depth + bid_depth
data = pd.DataFrame(columns=columns)
print(data)
```

```python
for filename in tqdm(filenames):
    try:
        with open(path + filename, 'r') as f:
            temp = json.load(f)
            for key in temp:
                ask = np.array(temp[key]['a'], dtype=float)
                bid = np.array(temp[key]['b'], dtype=float)
                ask_vol = ask.sum(axis=0)[1]
                bid_vol = bid.sum(axis=0)[1]
                ask_price = ask.min(axis=0)[0]
                bid_price = bid.max(axis=0)[0]
                date = datetime.fromtimestamp(int(key) / 1000)
                
                sorted_ask = ask[ask[:,0].argsort()]
                start = sorted_ask[0][0]
                ask_depth = np.zeros(20)
                diff = 10
                current_vol = 0
                cnt = 0
                for i in range(len(sorted_ask)):
                    if abs(start - sorted_ask[i][0]) < diff:
                        current_vol += sorted_ask[i][1]
                    else:
                        if cnt < 20:
                            ask_depth[cnt] = (current_vol)
                            cnt+=1
                            current_vol = 0
                            start = sorted_ask[i][0]
                        else:
                            continue

                
                sorted_bid = bid[bid[:,0].argsort()]
                start = sorted_bid[0][0]
                bid_depth = np.zeros(20)
                diff = 10
                current_vol = 0
                cnt = 0
                for i in range(len(sorted_bid)):
                    if abs(start - sorted_bid[i][0]) < diff:
                        current_vol += sorted_bid[i][1]
                    else:
                        if cnt < 20:
                            bid_depth[cnt] = (current_vol)
                            cnt+=1
                            current_vol = 0
                            start = sorted_bid[i][0]
                        else:
                            continue
                
                row = [date, int(key), ask_price, bid_price, ask_vol, bid_vol] + list(ask_depth) + list(bid_depth)
                data.loc[len(data)] = row
    except:
        print('Error at ' + str(filename))
        pass
```

```python
data['mid'] = (data['ask'] + data['bid']) / 2
data['change'] = data['mid'].pct_change()
```
