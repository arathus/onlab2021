```python
X = np.load('data/X.npy')
y = np.load('data/y.npy')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=False, random_state=42)
X_train = X_train.reshape(-1, 60, 44, 1)
X_test = X_test.reshape(-1, 60, 44, 1)
y_train = np.asarray(y_train).astype('float32').reshape((-1,1))
y_test = np.asarray(y_test).astype('float32').reshape((-1,1))
```

```python
sample_shape = X_train[0].shape
img_width, img_height = sample_shape[0], sample_shape[1]
input_shape = (img_width, img_height, 1)
```

```python
KERNEL_SIZE = (5, 5)
POOL_SIZE = (2, 2)
```

```python
input_img = Input(shape=input_shape)

x = Conv2D(16, (3, 3), activation='relu', padding='same')(input_img)
x = MaxPooling2D((4, 4), padding='same')(x)
x = Conv2D(8, (3, 3), activation='relu', padding='same')(x)
encoded = MaxPooling2D((2, 2), padding='same')(x)


x = Conv2D(8, (3, 3), activation='relu', padding='same')(encoded)
x = UpSampling2D((4, 4))(x)
x = Conv2D(16, (3, 3), activation='relu')(x)
x = UpSampling2D((2, 2))(x)
decoded = Conv2D(1, (3, 3), activation='linear', padding='same')(x)
```

```python
autoencoder = keras.Model(input_img, decoded)
autoencoder.compile(optimizer='adam', loss='mse')
autoencoder.summary()
```

```python
from keras.utils import plot_model

plot_model(
    autoencoder,
    to_file="model.png",
    show_shapes=False,
    show_layer_names=True,
    rankdir="TB",
    expand_nested=False,
    dpi=96,
)
```

```python
from keras.callbacks import TensorBoard

log_dir = '/tmp/autoencoder/' + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")

autoencoder.fit(X_train, X_train,
                epochs=30,
                batch_size=512,
                shuffle=False,
                validation_data=(X_test, X_test),
                callbacks=[TensorBoard(log_dir=log_dir)])
```

```python
pred = autoencoder.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
from tqdm.auto import tqdm
```

```python
# ugly code I know
mses = []
for i in tqdm(range(len(pred))):
    m = tf.keras.metrics.MeanSquaredError()
    m.update_state(X_test[i], pred[i])
    mse = m.result().numpy()
    mses.append(mse)
```

```python
mses = np.array(mses)
mses.mean(), mses.max(), mses.min()
```
